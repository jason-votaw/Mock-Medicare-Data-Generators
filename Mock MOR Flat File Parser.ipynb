```python
import pandas as pd

# HCCs with coefficients and descriptions (from uploaded image)
HCC_LIST = [
    {"code": "CNA_HCC1", "coefficient": 0.301, "description": "HIV/AIDS"},
    {"code": "CNA_HCC2", "coefficient": 0.500, "description": "Septicemia/Sepsis/Systemic Inflammatory Response Syndrome/Shock"},
    {"code": "CNA_HCC6", "coefficient": 0.381, "description": "Opportunistic Infections"},
    {"code": "CNA_HCC17", "coefficient": 4.209, "description": "Cancer Metastatic to Lung/Liver/Brain/and Other Organs; Acute Myeloid Leukemia Except Promyelocytic"},
    {"code": "CNA_HCC18", "coefficient": 2.341, "description": "Cancer Metastatic to Bone/Other and Unspecified Metastatic Cancer; Acute Leukemia Except Myeloid"},
    {"code": "CNA_HCC19", "coefficient": 1.798, "description": "Myelodysplastic Syndromes/Multiple Myeloma/and Other Cancers"},
    {"code": "CNA_HCC20", "coefficient": 1.136, "description": "Lung and Other Severe Cancers"},
    {"code": "CNA_HCC21", "coefficient": 0.671, "description": "Lymphoma and Other Cancers"},
    {"code": "CNA_HCC22", "coefficient": 0.363, "description": "Bladder/Colorectal/and Other Cancers"},
    {"code": "CNA_HCC23", "coefficient": 0.186, "description": "Prostate/Breast/and Other Cancers and Tumors"},
    {"code": "CNA_HCC35", "coefficient": 0.949, "description": "Pancreas Transplant Status"},
    {"code": "CNA_HCC36", "coefficient": 0.166, "description": "Diabetes with Severe Acute Complications"},
    {"code": "CNA_HCC37", "coefficient": 0.166, "description": "Diabetes with Chronic Complications"},
    {"code": "CNA_HCC38", "coefficient": 0.166, "description": "Diabetes with Glycemic/Unspecified/No Complications"},
    {"code": "CNA_HCC48", "coefficient": 0.186, "description": "Morbid Obesity"},
    {"code": "CNA_HCC49", "coefficient": 9.256, "description": "Specified Lysosomal Storage Disorders"},
    {"code": "CNA_HCC50", "coefficient": 0.648, "description": "Amyloidosis/Porphyria/and Other Specified Metabolic Disorders"},
    {"code": "CNA_HCC51", "coefficient": 0.510, "description": "Addisons and Cushings Diseases/Acromegaly/and Other Specified Endocrine Disorders"},
    {"code": "CNA_HCC62", "coefficient": 0.376, "description": "Liver Transplant Status/Complications"},
    {"code": "CNA_HCC63", "coefficient": 0.962, "description": "Chronic Liver Failure/End-Stage Liver Disorders"},
    {"code": "CNA_HCC64", "coefficient": 0.447, "description": "Cirrhosis of Liver"},
    {"code": "CNA_HCC65", "coefficient": 0.185, "description": "Chronic Hepatitis"},
    {"code": "CNA_HCC68", "coefficient": 0.388, "description": "Cholangitis and Obstruction of Bile Duct Without Gallstones"},
    {"code": "CNA_HCC77", "coefficient": 1.172, "description": "Intestine Transplant Status/Complications"},
    {"code": "CNA_HCC78", "coefficient": 0.326, "description": "Intestinal Obstruction/Perforation"},
    {"code": "CNA_HCC79", "coefficient": 0.357, "description": "Chronic Pancreatitis"},
    {"code": "CNA_HCC80", "coefficient": 0.550, "description": "Crohns Disease (Regional Enteritis)"},
    {"code": "CNA_HCC81", "coefficient": 0.244, "description": "Ulcerative Colitis"},
    {"code": "CNA_HCC92", "coefficient": 0.479, "description": "Bone/Joint/Muscle/Severe Soft Tissue Infections/Necrosis"},
    {"code": "CNA_HCC93", "coefficient": 0.617, "description": "Rheumatoid Arthritis and Other Specified Inflammatory Rheumatic Disorders"},
    {"code": "CNA_HCC94", "coefficient": 0.268, "description": "Systemic Lupus Erythematosus and Other Specified Systemic Connective Tissue Disorders"},
    {"code": "CNA_HCC107", "coefficient": 0.457, "description": "Sickle Cell Anemia (Hb-SS) and Thalassemia Beta Zero"},
    {"code": "CNA_HCC108", "coefficient": 0.146, "description": "Sickle Cell Disorders/Except Sickle Cell Anemia (Hb-SS) and Thalassemia Beta Zero; Beta Thalassemia Major"},
    {"code": "CNA_HCC109", "coefficient": 1.144, "description": "Acquired Hemolytic/Aplastic/and Sideroblastic Anemias"},
    {"code": "CNA_HCC111", "coefficient": 4.639, "description": "Hemophilia/Male"},
    {"code": "CNA_HCC112", "coefficient": 0.450, "description": "Immune Thrombocytopenia and Specified Coagulation Defects and Hemorrhagic Conditions"},
    {"code": "CNA_HCC114", "coefficient": 2.262, "description": "Common Variable and Combined Immunodeficiencies"},
    {"code": "CNA_HCC115", "coefficient": 0.565, "description": "Specified Immunodeficiencies and White Blood Cell Disorders"},
    {"code": "CNA_HCC125", "coefficient": 0.341, "description": "Dementia/Severe"},
    {"code": "CNA_HCC126", "coefficient": 0.341, "description": "Dementia/Moderate"},
    {"code": "CNA_HCC127", "coefficient": 0.341, "description": "Dementia/Mild or Unspecified"},
    {"code": "CNA_HCC135", "coefficient": 0.424, "description": "Drug Use with Psychotic Complications"},
    {"code": "CNA_HCC136", "coefficient": 0.424, "description": "Alcohol Use with Psychotic Complications"},
    {"code": "CNA_HCC137", "coefficient": 0.424, "description": "Drug Use Disorder/Moderate/Severe/Drug Use with Non-Psychotic Complications"},
    {"code": "CNA_HCC138", "coefficient": 0.423, "description": "Drug Use Disorder/Mild/Uncomplicated/Except Cannabis"},
    {"code": "CNA_HCC139", "coefficient": 0.242, "description": "Alcohol Use Disorder/Moderate/Severe/Alcohol Use with Specified Non-Psychotic Complications"},
    {"code": "CNA_HCC151", "coefficient": 0.511, "description": "Schizophrenia"},
    {"code": "CNA_HCC152", "coefficient": 0.484, "description": "Psychosis/Except Schizophrenia"},
    {"code": "CNA_HCC153", "coefficient": 0.396, "description": "Personality Disorders; Anorexia/Bulimia Nervosa"},
    {"code": "CNA_HCC154", "coefficient": 0.351, "description": "Bipolar Disorders without Psychosis"},
    {"code": "CNA_HCC155", "coefficient": 0.299, "description": "Major Depression/Moderate or Severe/without Psychosis"},
    {"code": "CNA_HCC180", "coefficient": 1.125, "description": "Quadriplegia"},
    {"code": "CNA_HCC181", "coefficient": 0.942, "description": "Paraplegia"},
    {"code": "CNA_HCC182", "coefficient": 0.478, "description": "Spinal Cord Disorders/Injuries"},
    {"code": "CNA_HCC190", "coefficient": 1.175, "description": "Amyotrophic Lateral Sclerosis and Other Motor Neuron Disease/Spinal Muscular Atrophy"},
    {"code": "CNA_HCC191", "coefficient": 0.855, "description": "Quadriplegic Cerebral Palsy"},
    {"code": "CNA_HCC192", "coefficient": 0.314, "description": "Cerebral Palsy/Except Quadriplegic"},
    {"code": "CNA_HCC193", "coefficient": 1.692, "description": "Chronic Inflammatory Demyelinating Polyneuritis and Multifocal Motor Neuropathy"},
    {"code": "CNA_HCC195", "coefficient": 2.909, "description": "Myasthenia Gravis with (Acute) Exacerbation"},
    {"code": "CNA_HCC196", "coefficient": 0.516, "description": "Myasthenia Gravis without (Acute) Exacerbation and Other Myoneural Disorders"},
    {"code": "CNA_HCC197", "coefficient": 0.426, "description": "Muscular Dystrophy"},
    {"code": "CNA_HCC198", "coefficient": 0.647, "description": "Multiple Sclerosis"},
    {"code": "CNA_HCC199", "coefficient": 0.615, "description": "Parkinson and Other Degenerative Disease of Basal Ganglia"},
    {"code": "CNA_HCC200", "coefficient": 0.279, "description": "Friedreich and Other Hereditary Ataxias; Huntington Disease"},
    {"code": "CNA_HCC201", "coefficient": 0.245, "description": "Seizure Disorders and Convulsions"},
    {"code": "CNA_HCC202", "coefficient": 0.543, "description": "Coma/Brain Compression/Anoxic Damage"},
    {"code": "CNA_HCC211", "coefficient": 0.879, "description": "Respirator Dependence/Tracheostomy Status/Complications"},
    {"code": "CNA_HCC212", "coefficient": 0.370, "description": "Respiratory Arrest"},
    {"code": "CNA_HCC213", "coefficient": 0.370, "description": "Cardio-Respiratory Failure and Shock"},
    {"code": "CNA_HCC221", "coefficient": 1.053, "description": "Heart Transplant Status/Complications"},
    {"code": "CNA_HCC222", "coefficient": 2.505, "description": "End-Stage Heart Failure"},
    {"code": "CNA_HCC223", "coefficient": 2.505, "description": "Heart Failure with Heart Assist Device/Artificial Heart"},
    {"code": "CNA_HCC224", "coefficient": 0.360, "description": "Acute on Chronic Heart Failure"},
    {"code": "CNA_HCC225", "coefficient": 0.360, "description": "Acute Heart Failure (Excludes Acute on Chronic)"},
    {"code": "CNA_HCC226", "coefficient": 0.360, "description": "Heart Failure/Except End-Stage and Acute"},
    {"code": "CNA_HCC227", "coefficient": 0.189, "description": "Cardiomyopathy/Myocarditis"},
    {"code": "CNA_HCC228", "coefficient": 0.252, "description": "Acute Myocardial Infarction"},
    {"code": "CNA_HCC229", "coefficient": 0.240, "description": "Unstable Angina and Other Acute Ischemic Heart Disease"},
    {"code": "CNA_HCC238", "coefficient": 0.299, "description": "Specified Heart Arrhythmias"},
    {"code": "CNA_HCC248", "coefficient": 0.239, "description": "Intracranial Hemorrhage"},
    {"code": "CNA_HCC249", "coefficient": 0.239, "description": "Ischemic or Unspecified Stroke"},
    {"code": "CNA_HCC253", "coefficient": 0.387, "description": "Hemiplegia/Hemiparesis"},
    {"code": "CNA_HCC254", "coefficient": 0.321, "description": "Monoplegia/Other Paralytic Syndromes"},
    {"code": "CNA_HCC263", "coefficient": 1.118, "description": "Atherosclerosis of Arteries of the Extremities with Ulceration or Gangrene"},
    {"code": "CNA_HCC264", "coefficient": 0.455, "description": "Vascular Disease with Complications"},
    {"code": "CNA_HCC267", "coefficient": 0.294, "description": "Deep Vein Thrombosis and Pulmonary Embolism"},
    {"code": "CNA_HCC276", "coefficient": 2.531, "description": "Lung Transplant Status/Complications"},
    {"code": "CNA_HCC277", "coefficient": 0.998, "description": "Cystic Fibrosis"},
    {"code": "CNA_HCC278", "coefficient": 0.818, "description": "Idiopathic Pulmonary Fibrosis and Lung Involvement in Systemic Sclerosis"},
    {"code": "CNA_HCC279", "coefficient": 0.818, "description": "Severe Persistent Asthma"},
    {"code": "CNA_HCC280", "coefficient": 0.319, "description": "Chronic Obstructive Pulmonary Disease/Interstitial Lung Disorders/and Other Chronic Lung Disorders"},
    {"code": "CNA_HCC282", "coefficient": 0.440, "description": "Aspiration and Specified Bacterial Pneumonias"},
    {"code": "CNA_HCC283", "coefficient": 0.204, "description": "Empyema/Lung Abscess"},
    {"code": "CNA_HCC298", "coefficient": 0.336, "description": "Severe Diabetic Eye Disease/Retinal Vein Occlusion/and Vitreous Hemorrhage"},
    {"code": "CNA_HCC300", "coefficient": 0.596, "description": "Exudative Macular Degeneration"},
    {"code": "CNA_HCC326", "coefficient": 0.815, "description": "Chronic Kidney Disease/Stage 5"},
    {"code": "CNA_HCC327", "coefficient": 0.514, "description": "Chronic Kidney Disease/Severe (Stage 4)"},
    {"code": "CNA_HCC328", "coefficient": 0.127, "description": "Chronic Kidney Disease/Moderate (Stage 3B)"},
    {"code": "CNA_HCC329", "coefficient": 0.127, "description": "Chronic Kidney Disease/Moderate (Stage 3/Except 3B)"},
    {"code": "CNA_HCC379", "coefficient": 1.965, "description": "Pressure Ulcer of Skin with Necrosis Through to Muscle/Tendon/Bone"},
    {"code": "CNA_HCC380", "coefficient": 1.078, "description": "Chronic Ulcer of Skin/Except Pressure/Through to Bone or Muscle"},
    {"code": "CNA_HCC381", "coefficient": 1.075, "description": "Pressure Ulcer of Skin with Full Thickness Skin Loss"},
    {"code": "CNA_HCC382", "coefficient": 0.838, "description": "Pressure Ulcer of Skin with Partial Thickness Skin Loss"},
    {"code": "CNA_HCC383", "coefficient": 0.646, "description": "Chronic Ulcer of Skin/Except Pressure/Not Specified as Through to Bone or Muscle"},
    {"code": "CNA_HCC385", "coefficient": 1.291, "description": "Severe Skin Burn"},
    {"code": "CNA_HCC387", "coefficient": 0.406, "description": "Pemphigus/Pemphigoid/and Other Specified Autoimmune Skin Disorders"},
    {"code": "CNA_HCC397", "coefficient": 0.199, "description": "Major Head Injury with Loss of Consciousness > 1 Hour"},
    {"code": "CNA_HCC398", "coefficient": 0.199, "description": "Major Head Injury with Loss of Consciousness < 1 Hour or Unspecified"},
    {"code": "CNA_HCC399", "coefficient": 0.199, "description": "Major Head Injury without Loss of Consciousness"},
    {"code": "CNA_HCC401", "coefficient": 0.522, "description": "Vertebral Fractures without Spinal Cord Injury"},
    {"code": "CNA_HCC402", "coefficient": 0.467, "description": "Hip Fracture/Dislocation"},
    {"code": "CNA_HCC405", "coefficient": 0.598, "description": "Traumatic Amputations and Complications"},
    {"code": "CNA_HCC409", "coefficient": 0.598, "description": "Amputation Status/Lower Limb/Amputation Complications"},
    {"code": "CNA_HCC454", "coefficient": 1.068, "description": "Stem Cell/Including Bone Marrow/Transplant Status/Complications"},
    {"code": "CNA_HCC463", "coefficient": 0.673, "description": "Artificial Openings for Feeding or Elimination"}
]

# Extract only the codes for flagging in the fixed-width file
HCC_POOL = [entry["code"] for entry in HCC_LIST]

# Demographic score lookup table based on age and sex
DEMOGRAPHIC_SCORES = {
    # Female scores
    ('F', '1'): 0.330,  # Female 65-69
    ('F', '2'): 0.395,  # Female 70-74
    ('F', '3'): 0.465,  # Female 75-79
    ('F', '4'): 0.524,  # Female 80-84
    ('F', '5'): 0.624,  # Female 85-89
    ('F', '6'): 0.737,  # Female 90-94
    ('F', '7'): 0.742,  # Female 95+
    # Male scores
    ('M', '1'): 0.332,  # Male 65-69
    ('M', '2'): 0.396,  # Male 70-74
    ('M', '3'): 0.502,  # Male 75-79
    ('M', '4'): 0.571,  # Male 80-84
    ('M', '5'): 0.664,  # Male 85-89
    ('M', '6'): 0.664,  # Male 90-94 (using 85-89 score)
    ('M', '7'): 0.664   # Male 95+ (using 85-89 score)
}

def get_demographic_score_from_indicators(sex, age_group_digit):
    """Get demographic score based on sex and age group digit from flat file"""
    return DEMOGRAPHIC_SCORES.get((sex, age_group_digit), 0.330)  # Default to female 65-69 if not found

def parse_mor_flatfile(file_path, output_csv):
    records = []

    with open(file_path, 'r') as f:
        for line_num, line in enumerate(f, 1):
            line = line.strip()  # Remove newline characters
            
            # Skip header record (line starting with "1H")
            if line.startswith("1H"):
                print(f"Skipping header record: {line[:20]}...")
                continue
                
            # Skip if line is too short
            if len(line) < 100:
                print(f"Skipping short line {line_num}: {len(line)} characters")
                continue
            
            try:
                # Parse based on the actual structure of the generated file
                # Structure: RECORD_TYPE(1) + MBI(11) + LAST_NAME(12) + FIRST_NAME(12) + MI(1) + MI(1) + DOB(8) + SEX_DIGIT(1) + AGE_DIGIT(1) + PADDING(20) + HCC_FLAGS(103) + PADDING(30)
                
                record = {
                    "RECORD_TYPE": line[0],                    # Position 0
                    "MBI": line[1:12].strip(),                 # Positions 1-11
                    "LAST_NAME": line[12:24].strip(),          # Positions 12-23
                    "FIRST_NAME": line[24:36].strip(),         # Positions 24-35
                    "MI": line[36],                            # Position 36
                    "DOB": line[38:46],                        # Positions 38-45
                    "SEX_INDICATOR": line[46],                 # Position 46: 1=Male, 2=Female
                    "AGE_GROUP_INDICATOR": line[47],           # Position 47: 1-7 age groups
                }
                
                # Calculate demographic score based on sex and age group indicators
                sex_char = 'M' if record["SEX_INDICATOR"] == '1' else 'F'
                calculated_demo_score = get_demographic_score_from_indicators(sex_char, record["AGE_GROUP_INDICATOR"])
                record["DEMOGRAPHIC_SCORE"] = calculated_demo_score
                
                # Convert sex indicator to readable format
                record["SEX"] = sex_char
                
                # Convert age group indicator to readable format
                age_groups = {
                    '1': '65-69', '2': '70-74', '3': '75-79', '4': '80-84', 
                    '5': '85-89', '6': '90-94', '7': '95+'
                }
                record["AGE_GROUP"] = age_groups.get(record["AGE_GROUP_INDICATOR"], "Unknown")
                
                # Parse HCC flags starting from position 68 (after 20 chars of demographic padding)
                hcc_start_pos = 68
                total_disease_score = 0
                active_hccs = []
                
                # Create HCC coefficient lookup dictionary for faster access
                hcc_coeff_lookup = {entry["code"]: entry["coefficient"] for entry in HCC_LIST}
                
                for i, hcc in enumerate(HCC_POOL):
                    if hcc_start_pos + i < len(line):
                        hcc_flag = int(line[hcc_start_pos + i])
                        if hcc_flag == 1:
                            # Populate with coefficient instead of 1
                            coefficient = hcc_coeff_lookup.get(hcc, 0)
                            record[hcc] = coefficient
                            total_disease_score += coefficient
                            active_hccs.append(hcc)
                        else:
                            record[hcc] = 0.0  # Use 0.0 for consistency with coefficients
                    else:
                        record[hcc] = 0.0  # Default if position doesn't exist
                        
                record["DISEASE_SCORE"] = round(total_disease_score, 4)
                record["ACTIVE_HCC_COUNT"] = len(active_hccs)
                record["ACTIVE_HCCS"] = ", ".join(active_hccs) if active_hccs else "None"
                
                # Calculate total RAF (Risk Adjustment Factor) = demographic + disease score
                record["TOTAL_RAF"] = round(record["DEMOGRAPHIC_SCORE"] + record["DISEASE_SCORE"], 4)
                
                records.append(record)
                
            except Exception as e:
                print(f"Error parsing line {line_num}: {e}")
                print(f"Line content: {line[:50]}...")
                continue

    df = pd.DataFrame(records)
    
    if df.empty:
        print("No records were successfully parsed!")
        return
    
    # Display first 18 columns of the CSV output as preview
    print("=" * 120)
    print("FIRST 18 COLUMNS OF CSV OUTPUT (PREVIEW):")
    print("=" * 120)
    
    # Show first 18 column headers
    first_18_columns = df.columns[:18].tolist()
    print("First 18 column headers:")
    for i, col in enumerate(first_18_columns, 1):
        print(f"{i:2d}. {col}")
    
    # Display first 10 records with only the first 18 columns
    print("\n" + "=" * 120)
    print("FIRST 10 RECORDS (FIRST 18 COLUMNS ONLY):")
    print("=" * 120)
    
    # Set pandas options for better display
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    pd.set_option('display.max_colwidth', 15)
    
    print(df[first_18_columns].head(10).to_string(index=False))
    
    print("\n" + "=" * 120)
    print("HCC SUMMARY:")
    print("=" * 120)
    
    # Show summary statistics
    print(f"Total records processed: {len(df)}")
    print(f"Records with HCCs: {df[df['ACTIVE_HCC_COUNT'] > 0].shape[0]}")
    print(f"Records without HCCs: {df[df['ACTIVE_HCC_COUNT'] == 0].shape[0]}")
    print(f"Average HCCs per member: {df['ACTIVE_HCC_COUNT'].mean():.2f}")
    print(f"Average demographic score: {df['DEMOGRAPHIC_SCORE'].mean():.4f}")
    print(f"Average disease score: {df['DISEASE_SCORE'].mean():.4f}")
    print(f"Average total RAF: {df['TOTAL_RAF'].mean():.4f}")
    print(f"Total columns: {len(df.columns)}")
    print(f"Showing first 18 columns out of {len(df.columns)} total columns")
    
    # Show top HCCs by total coefficient value
    hcc_totals = {}
    for hcc in HCC_POOL:
        hcc_totals[hcc] = df[hcc].sum()
    
    top_hccs = sorted(hcc_totals.items(), key=lambda x: x[1], reverse=True)[:10]
    print("\nTop 10 HCCs by total coefficient value:")
    for hcc, total_coeff in top_hccs:
        if total_coeff > 0:
            count = (df[hcc] > 0).sum()
            avg_coeff = total_coeff / count if count > 0 else 0
            print(f"  {hcc}: {total_coeff:.4f} total ({count} members, avg {avg_coeff:.4f})")
    
    # Show RAF distribution
    print(f"\nRAF Score Distribution:")
    print(f"  Min RAF: {df['TOTAL_RAF'].min():.4f}")
    print(f"  Max RAF: {df['TOTAL_RAF'].max():.4f}")
    print(f"  Median RAF: {df['TOTAL_RAF'].median():.4f}")
    
    print("=" * 120)
    
    # Save to CSV
    df.to_csv(output_csv, index=False)
    print(f"\nCSV saved to: {output_csv}")

# Example usage
parse_mor_flatfile("mor_mock_flatfile.txt", "parsed_mor_flatfile.csv")
```
